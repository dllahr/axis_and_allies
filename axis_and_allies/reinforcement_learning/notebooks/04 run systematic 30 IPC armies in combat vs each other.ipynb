```python
import pprint
import time

import pandas
import numpy

import plotly.express as pltxpr

import axis_and_allies.reinforcement_learning.super_simple_env as super_simple_env
import axis_and_allies.run_simulation as run_simulation
import axis_and_allies.combat as combat
```

```python
input_armies = "possible_armies_IPC_30_r66x6.txt"
input_unit_dict = "../../unit_data.json"
N_sim = 1000
```

```python
output_template = "output/data/30_IPC_army/all_result_metrics_r{}x{}.txt"
```

```python
unit_dict = run_simulation.load_units(input_unit_dict)
pprint.pprint(unit_dict)
```

```python
armies_df = pandas.read_csv(input_armies, sep="\t")
armies_df
```

```python
start_time = time.time()
prev_time = start_time

all_result_list = []
for attack_ind in range(armies_df.shape[0]):
    attack_row = armies_df.loc[attack_ind].to_numpy()
#     print("################ attack_row:", attack_ind, attack_row)
    
    attack_unit_names = super_simple_env.convert_unit_count_to_unit_name_list(attack_row[:-1])
#     print("attack_unit_names", attack_unit_names)
    
    for defense_ind in range(armies_df.shape[0]):
        defense_row = armies_df.loc[defense_ind].to_numpy()
#         print("################ defense_row:", defense_ind, defense_row)
        
        defense_unit_names = super_simple_env.convert_unit_count_to_unit_name_list(defense_row[:-1])
#         print("defense_unit_names", defense_unit_names)

        cur_result_list = []
        all_result_list.append((attack_ind, defense_ind, cur_result_list))

        for i in range(N_sim):
            result = run_simulation.run_single_from_names(
                unit_dict, attack_unit_names, defense_unit_names, combat.BATTLE_TYPE_LAND
            )
    #         print(result[-1])
            cur_result_list.append(result[-1])
    #         break

        cur_time = time.time()
        if (cur_time - prev_time) >= 10.:
#             print("progress:", i)
            print("################ attack_row:", attack_ind, attack_row)
            print("################ defense_row:", defense_ind, defense_row)
            prev_time = cur_time
            
#     break

end_time = time.time()
duration = end_time - start_time
print("duration:", duration, duration/60.)

print(len(all_result_list))
```

```python
all_result_metric_list = []

prev_attack_ind = -1
for attack_ind, defense_ind, cur_result_list in all_result_list:
    if attack_ind != prev_attack_ind:
        print("##################", attack_ind, defense_ind)
        prev_attack_ind = attack_ind

    sum_start_ipc_attack = armies_df.loc[attack_ind, "cost"]
    sum_start_ipc_defense = armies_df.loc[defense_ind, "cost"]
#     print(sum_start_ipc_attack, sum_start_ipc_defense)
    
    cur_result_metric_list = []
    all_result_metric_list.append((attack_ind, defense_ind, cur_result_metric_list))
    for cur_result in cur_result_list:
#         print(cur_result)
        cur_metric = run_simulation.calculate_metrics_from_combat_result(
            cur_result, sum_start_ipc_attack, sum_start_ipc_defense
        )
#         print(cur_metric)
        cur_result_metric_list.append(cur_metric)

print(len(all_result_metric_list))
print(set([len(x[2]) for x in all_result_metric_list]))
print(set([len(x[2]) for x in all_result_metric_list]))
```

```python
t = []
for attack_ind, defense_ind, cur_result_metric_list in all_result_metric_list:
    df = pandas.DataFrame({
        "attack_army_id":attack_ind,
        "defense_army_id":defense_ind,
        "fraction_ipc_winner":[x.fraction_ipc_winner for x in cur_result_metric_list]
    })
    t.append(df)

result_df = pandas.concat(t, axis=0)
# result_df["army_id_str"] = result_df.army_id.astype(str)
result_df
```

```python
output_file = output_template.format(result_df.shape[0], result_df.shape[1])
print(output_file)
result_df.to_csv(output_file, sep="\t", index=False)
```
