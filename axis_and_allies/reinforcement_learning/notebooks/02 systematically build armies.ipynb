```python
import numpy
import pandas
```

```python
output_template = "possible_armies_IPC_30_r{}x{}.txt"
```

```python
def recurse_build_by_count(total_count, current_entry, all_entry_set):
    if sum(current_entry) == total_count:
        all_entry_set.add(tuple(current_entry))
    else:
        for i in range(len(current_entry)):
            new_entry = list(current_entry)
            new_entry[i] += 1
            
            recurse_build_by_count(total_count, new_entry, all_entry_set)
            
ae_set = set()
recurse_build_by_count(4, [0,0,0], ae_set)
ae_list = sorted(ae_set)
print(len(ae_list))
print(ae_list)
```

```python
def recurse_build_by_IPC(IPC_cost_arr, total_IPC, current_entry, all_entry_set):
    entry_cost = sum(IPC_cost_arr * current_entry)
    
    for i,cost  in enumerate(IPC_cost_arr):
        if entry_cost + cost > total_IPC:
            all_entry_set.add(tuple(current_entry))
        else:
            new_entry = numpy.array(current_entry)
            new_entry[i] += 1
            recurse_build_by_IPC(IPC_cost_arr, total_IPC, new_entry, all_entry_set)
        
    

IPC_cost_arr = numpy.array([3,4,6,10,12])
init_entry = numpy.zeros(IPC_cost_arr.shape[0])
ae_set = set()
recurse_build_by_IPC(IPC_cost_arr, 30, init_entry, ae_set)
ae_list = sorted(ae_set)
print(len(ae_list))
ae_list
```

```python
t = numpy.array(ae_list)
# print(t)

army_df = pandas.DataFrame(t, columns=["inf", "art", "tank", "fight", "bomb"], dtype=int)
army_df["cost"] = (army_df * IPC_cost_arr).sum(axis=1)
army_df
```

```python
t = army_df.loc[army_df.cost <= 27]
display(t)

remain_df = army_df.loc[army_df.cost > 27]
# display(remain.loc[remain.bomb == 2])

for ind in t.index:
#     print("############# {}".format(ind))
    row = t.loc[ind]
#     display(t.loc[[ind]])
    u = row.values[:-1] <= remain_df.iloc[:,:-1]
#     display(u)
    v = u.all(axis=1)
#     print(v)
#     display(remain_df.loc[v])
    
    if not any(v):
        print("did not find a better row for ind:  {}".format(ind))
```
